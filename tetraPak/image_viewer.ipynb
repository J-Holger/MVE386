```python
proposal = 20240661
visit = 2024102408
folder = '/data/visitors/formax/%s/%s/process' % (proposal, visit)

logfile = 'logbook_227.csv'
df = pd.read_csv(os.path.join(folder,logfile), sep=';')
df

# drop empty columns
df = df.dropna(axis=1, how='all')
df
```

```python
def show_image(scan_nr):
    image_folder = os.path.join(folder, 'images')
    filename = 'tot_scatter_scan_%i.png' % scan_nr
    print(filename)
    img = plt.imread(os.path.join(folder, filename))
    plt.imshow(img)
    plt.show()

interact(show_image, scan_nr=(0, 100, 1));
```
